```python
import pandas
import plotly.express as px
from colab.CA3 import utils
```

```python
!ls colab/CA3
```

```python
df = pandas.read_csv("colab/CA3/frequencies.csv")
df = df.drop(['Unnamed: 0'], axis=1)
print(df.columns)
print(df)
```

```python
df['MSE'] = df.apply(utils.mse, axis=1)
print(df['MSE'])
```

```python
fig = px.scatter_3d(
    df,
    x='netParams.connParams.PYR->BC_AMPA.weight',
    y='netParams.connParams.PYR->OLM_AMPA.weight',
    z='netParams.connParams.PYR->PYR_AMPA.weight',
    color='MSE'
)

fig.show()
```

```python
UB = 1
filtered_df = df[df['MSE'] < UB]
fig = px.scatter_3d(
    filtered_df,
    x='netParams.connParams.PYR->BC_AMPA.weight',
    y='netParams.connParams.PYR->OLM_AMPA.weight',
    z='netParams.connParams.PYR->PYR_AMPA.weight',
    color='MSE'
)

fig.show()
```

```python
filtered_mse = df[df['MSE'] < 0.5]

fig = px.scatter_3d(
    filtered_mse,
    x='netParams.connParams.PYR->BC_AMPA.weight',
    y='netParams.connParams.PYR->OLM_AMPA.weight',
    z='netParams.connParams.PYR->PYR_AMPA.weight',
    color='MSE'
)

fig.show()
```
